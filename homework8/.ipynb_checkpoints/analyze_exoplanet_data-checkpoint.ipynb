```python
df = pd.read_csv('exoplanet_data.csv', comment='#')
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
# Table Label: Planet Name
# Database Column Name: pl_name
# Description: The name given to the exoplanet
# In my own words: The unique identifier/name of each discovered planet

# Table Label: Host Name
# Database Column Name: hostname
# Description: Name of the host star the planet orbits
# In my own words: The star that the exoplanet revolves around

# Table Label: Discovery Method
# Database Column Name: discoverymethod
# Description: Method by which the planet was discovered
# In my own words: The technique astronomers used to find the planet

# Table Label: Spectral Type
# Database Column Name: st_spectype
# Description: Classification of the host star based on its spectral properties
# In my own words: The type/category of the star based on its temperature and light

# Table Label: Equilibrium Temperature [K]
# Database Column Name: pl_eqt
# Description: Estimated equilibrium temperature of the planet in Kelvin
# In my own words: How hot the planet surface is estimated to be

# Table Label: Planet Mass or Mass*sin(i) [Jupiter Mass]
# Database Column Name: pl_bmassj
# Description: Planet mass measured in units of Jupiter's mass
# In my own words: How heavy the planet is compared to Jupiter
```

```python
df_small = df[['pl_name', 'hostname', 'discoverymethod', 
                'st_spectype', 'pl_eqt', 'pl_bmassj']]
print(df_small.shape)
```

```python
df_small = df_small.dropna()
print(df_small.shape)
```

```python
print(df_small.shape)
print(df_small.head())
print(df_small.tail())
print(df_small.columns.tolist())
```

```python
print(df_small['discoverymethod'].value_counts())
```

```python
# Transit: detecting a dip in starlight as a planet passes in front of its star
# Radial Velocity: detecting wobble in a star caused by a planet's gravity
# Imaging: directly photographing the planet
# Microlensing: detecting light bending from a distant star due to a planet's gravity
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

colors = {'Transit': 'blue', 'Radial Velocity': 'green', 
          'Imaging': 'red', 'Microlensing': 'orange'}

for method, group in df_small.groupby('discoverymethod'):
    ax.scatter(group['pl_bmassj'], group['pl_eqt'], 
               label=method, alpha=0.6,
               color=colors.get(method, 'purple'))

ax.set_title('Planet Mass vs Equilibrium Temperature')
ax.set_xlabel('Planet Mass (Jupiter Masses)')
ax.set_ylabel('Equilibrium Temperature (K)')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for method, group in df_small.groupby('discoverymethod'):
    ax.scatter(np.log(group['pl_bmassj']), np.log(group['pl_eqt']), 
               label=method, alpha=0.6,
               color=colors.get(method, 'purple'))

ax.set_title('Planet Mass vs Equilibrium Temperature (Log Scale)')
ax.set_xlabel('Log Planet Mass (Jupiter Masses)')
ax.set_ylabel('Log Equilibrium Temperature (K)')
ax.legend()
plt.show()
```

```python
# Based on the logarithmic plot, exoplanets found by the imaging method 
# tend to be MORE massive since they cluster toward the higher end of the 
# mass axis. This makes sense because imaging works better for large, 
# bright planets that are easier to photograph directly.
```

```python
filtered = df_small[df_small['pl_bmassj'] > 5]
coldest = filtered.loc[filtered['pl_eqt'].idxmin()]
print(coldest)
```

```python
print("Name:", coldest['pl_name'])
print("Index:", coldest.name)
print("Equilibrium Temperature:", coldest['pl_eqt'], "K")
print("Mass:", coldest['pl_bmassj'], "Jupiter Masses")
print("Spectral Type:", coldest['st_spectype'])
```

```python
# Is it a sun-like star?
# The Sun is a G2 V type star. eps Ind A is a K5 V star, which is slightly 
# cooler and smaller than the Sun but in the same general category of 
# main-sequence stars. It is not exactly sun-like but is a similar type of star
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for method, group in df_small.groupby('discoverymethod'):
    ax.scatter(np.log(group['pl_bmassj']), np.log(group['pl_eqt']),
               label=method, alpha=0.6,
               color=colors.get(method, 'purple'))

# Highlight coldest planet
ax.scatter(np.log(coldest['pl_bmassj']), np.log(coldest['pl_eqt']),
           color='black', marker='*', s=300,
           label=coldest['pl_name'], zorder=5)

ax.set_title('Planet Mass vs Equilibrium Temperature (Log Scale)')
ax.set_xlabel('Log Planet Mass (Jupiter Masses)')
ax.set_ylabel('Log Equilibrium Temperature (K)')
ax.legend()
plt.show()
```
